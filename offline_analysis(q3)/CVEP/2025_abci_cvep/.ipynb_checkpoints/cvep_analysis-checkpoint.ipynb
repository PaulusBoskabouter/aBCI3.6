```python
import os

import mne
from mnelab_read_raw import read_raw_xdf as read_raw
import numpy as np
import pyxdf
import pyntbci
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_context("paper", font_scale=1.5)
```

```python
# Data settings
data_dir = os.path.join(os.path.expanduser("~"), "Downloads", "cvep_speller")
subject = 1
n_runs = 3

# Preprocessing settings
fs = 120  # Sample rate for downsampling (Hz)
tmin = 0  # Onset of trials (s)
tmax = 4.2  # Duration of trials (s)
l_freq = 6.0  # High pass filter cutoff (Hz)
h_freq = 30.0  # Low pass filter cutoff (Hz)

# Stimulus settings
fr = 60  # Stimulus presentation rate (Hz)
n_classes = 53  # The speller contained 52 symbols
```

```python
# Load stimuli
V = np.load(os.path.join(data_dir, "mgold_61_6521.npz"))["codes"].repeat(int(fs / fr), axis=1)[:n_classes, :]

# Print summary
print("Stimuli", V.shape, ("stimuli x samples"))

# Plot stimulus sequences
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
pyntbci.plotting.stimplot(V, fs=fr, ax=ax)
```

```python
# Load EEG data
X = []
y = []
for run in range(1, 1+n_runs):

    # Run file name
    fn = os.path.join(data_dir, f"sub-p{subject:03d}_ses-s001_run-{run:03d}_task-training.xdf")

    # Load marker stream
    streams = pyxdf.load_xdf(fn)[0]
    names = [stream["info"]["name"][0] for stream in streams]
    markers = streams[names.index("cvep-speller-stream")]
    
    # Extract trial labels
    y.append(np.array([
        int(marker[0].split(";")[1].split("=")[1]) 
        for marker in markers["time_series"] 
        if marker[0].startswith("start_cue")
    ]))

    # Extract trial onsets
    onsets = np.array([
        float(time) 
        for time, marker in zip(markers["time_stamps"], markers["time_series"])
        if marker[0].startswith("start_trial")
    ])

    # Find closest EEG timestamp to marker onsets
    events = np.zeros((len(onsets), 3), dtype="int")
    events[:, 0] = np.argmin(np.abs(streams[names.index("BioSemi")]["time_stamps"] - onsets[:, None]), axis=1)
    
    # Load EEG stream
    streams = pyxdf.resolve_streams(fn)
    names = [stream["name"] for stream in streams]
    stream_id = streams[names.index("BioSemi")]["stream_id"]
    raw = read_raw(fn, stream_ids=[stream_id])
    
    # Filtering
    raw = raw.filter(l_freq=l_freq, h_freq=h_freq, picks=np.arange(1, 65), verbose=False)
    
    # Slicing
    # N.B. add 0.5 sec pre and post trial to capture filter artefacts of anti-aliasing in downsampling (removed later on)
    epo = mne.Epochs(raw, events=events, tmin=tmin-0.5, tmax=tmax+0.5, baseline=None, picks="eeg", preload=True, verbose=False)
    
    # Resampling
    # N.B. Downsampling is done after slicing to maintain accurate stimulus timing
    epo = epo.resample(sfreq=fs, verbose=False)
    
    # Extract relevant EEG data
    X.append(epo.get_data(tmin=tmin, tmax=tmax))

# Stack all trials across runs
X = np.concatenate(X, axis=0)
y = np.concatenate(y, axis=0)

# Print summary of data
print("EEG data:", X.shape, "(trials x channels x samples)")
print("Labels:", y.shape, "(trials)")
```

```python
# Fit a classifier on all data to visualize responses
rcca = pyntbci.classifiers.rCCA(stimulus=V, fs=fs, event="contrast", onset_event=True, encoding_length=0.3)
rcca.fit(X, y)

# Plot spatial and temporal filter
fig, ax = plt.subplots(1, 2, figsize=(15, 3))
pyntbci.plotting.topoplot(rcca.w_, locfile=os.path.join(data_dir, "biosemi32.loc"), ax=ax[0])
ax[1].plot(np.arange(1/fs, 0.3+1/fs, 1/fs), rcca.r_.reshape((len(rcca.events_), -1)).T)
ax[1].legend(rcca.events_)
ax[1].set_xlabel("time [s]")
ax[1].set_ylabel("amplitde [au]")
```

```python
# Cross-validation with full trial
n_folds = 4
folds = np.arange(n_folds).repeat(int(np.ceil(X.shape[0] / n_folds)))[:X.shape[0]]

accuracy = np.zeros(n_folds)
for i_fold in range(n_folds):
    
    # Split data in train and test
    X_trn, y_trn = X[folds != i_fold, :, :], y[folds != i_fold]
    X_tst, y_tst = X[folds == i_fold, :, :], y[folds == i_fold]

    # Set up classifier
    rcca = pyntbci.classifiers.rCCA(stimulus=V, fs=fs, event="contrast", onset_event=True, encoding_length=0.3)

    # Train classifier
    rcca.fit(X_trn, y_trn)

    # Apply classifier
    yh_tst = rcca.predict(X_tst)

    # Evaluate performance
    accuracy[i_fold] = np.mean(yh_tst == y_tst)

print("Accuracy:", np.mean(accuracy))
    
```

```python
# Cross-validation with dynamic stopping
n_folds = 4
folds = np.arange(n_folds).repeat(int(np.ceil(X.shape[0] / n_folds)))[:X.shape[0]]

segment_time = 0.1
n_segments = int(X.shape[2] / (fs * segment_time))

accuracy = np.zeros((n_folds))
duration = np.zeros((n_folds))
for i_fold in range(n_folds):

    # Split data in train and test
    X_trn, y_trn = X[folds != i_fold, :, :], y[folds != i_fold]
    X_tst, y_tst = X[folds == i_fold, :, :], y[folds == i_fold]

    # Set up classifier
    rcca = pyntbci.classifiers.rCCA(stimulus=V, fs=fs, event="contrast", onset_event=True, encoding_length=0.3)
    stop = pyntbci.stopping.DistributionStopping(estimator=rcca, fs=fs, segment_time=segment_time, distribution="beta", target_p=0.999)

    # Train classifier
    stop.fit(X_trn, y_trn)

    # Apply classifier
    yh_tst = np.zeros(y_tst.size)
    yh_dur = np.zeros(y_tst.size)
    for i_trial in range(y_tst.size):
        for i_segment in range(n_segments):
            
            # Use data of the current trial and until the current segment only
            yh = stop.predict(X_tst[[i_trial], :, :int((1+i_segment) * segment_time * fs)])[0]

            # Stop classifying this trial if the classifier is sufficiently certain (if uncertain, yh=-1)
            if yh >= 0:
                yh_tst[i_trial] = yh
                yh_dur[i_trial] = (1 + i_segment) * segment_time
                break

    # Evaluate performance
    accuracy[i_fold] = np.mean(yh_tst == y_tst)
    duration[i_fold] = np.mean(yh_dur)

print("Accuracy:", np.mean(accuracy))
print("Duration:", np.mean(duration))
    
```
